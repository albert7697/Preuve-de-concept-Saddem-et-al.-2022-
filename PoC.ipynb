 <H3><i> Machine learning-based approach for online fault Diagnosis of Discrete Event System (Saddem et al. - 2022 )</H3> 

# Introduction

Dans dans le contexte de l'industrie du futur où les systèmes sont de plus en plus complexes, les systèmes automatisés de production, tels que les lignes de convoyage ou les systèmes de tri sont largement modélisés comme des Systèmes à Événements Discrets (SED). Dans ce type de systèmes, l’évolution dynamique est régi par des événements discrets (changements d’état de capteurs ou commutations d’actionneurs), plutôt que par des lois continues. Dans un environnement concurrentiel marqué par les exigences de disponibilité, de sécurité et de qualité, la capacité à détecter et diagnostiquer rapidement les défauts est devenue un enjeu majeur. Un défaut non détecté peut entraîner des arrêts non planifiés, des pertes économiques importantes, voire des situations dangereuses pour les opérateurs. Les approches classiques de diagnostic des SED (basée sur les modèles et ou basée sur la connaissance) présentent des limites majeures que sont : la nécessité d’un modèle analytique précis et complet du système, coûts élevés de développement et de maintenance, la difficulté à mettre à jour de nouveaux défauts, dépendance de la disponibilité d’experts, etc.
Avec l’essor de l’industrie 4.0 et la disponibilité croissante des données issues des capteurs industriels, les approches basées sur les données et en particulier celles basées sur le Machine Learning apparaissent comme une alternative prometteuse. Parmi elles, les réseaux de neurones récurrents, et plus spécifiquement les Long Short-Term Memory (LSTM), sont particulièrement adaptés à l’analyse de données temporelles et séquentielles.
C’est dans ce contexte que s’inscrit l’approche proposée par <B>Saddem et Baptiste (2022)</b> (<a href = "./Saddem et al. - 2022 - Machine learning-based approach for online fault Diagnosis of Discrete Event Sy.pdf"><b><i> Machine learning-based approach for online fault Diagnosis of Discrete Event System</i></b></a>), qui combine une modélisation des SED avec des techniques d’apprentissage automatique afin de réaliser un <b>diagnostic de défauts en ligne</b>, capable de distinguer plusieurs types de fautes à partir de séquences temporelles de signaux discrets. Notre objectif est de réaliser une preuve de concepts des travaux realisés par les auteurs de cet article, par une démonstration de la faisabilité ou non et de la cohérence méthodiologique. Pour cela, nous avons reparti notre travail en plusieurs parties: Premièrement, nous commençons par la modélisation du problème; deuxièmement, nous faisons l'acquisition des données; Troisiement, la préparation des données pour l'algorithme de RNN - LSTM; quatrièment, l'entrainement et la validation de l'algorithme et en fin, une évaluation des performances. 

# 1. Modélisation du problème

Le problème a été modélisé ainsi qu'il suit : <br>
Pour être cohérent avec l’article, nous avons considéré un système automatisé de production simplifié. Il s'agir d'un système de convoyage constitué de : 
* Deux capteurs binaires (Pour détecter la présence d'une pièce à l'entrée et à la sortie) nommés "capteur_1" et "capteur_2"
* Un actionneur binaire représentant le moteur du convoyeur nommé "moteur" <br>

Le comportement nominal du système est cyclique : Une pièce est détectée par capteur_1, transportée par le convoyeur (moteur), puis, détectée par capteur_2. A partir de ce comportement normal, plusieurs types de défauts ont été définis : 
* Capteur bloqué sur 0 ou sur 1
* Moteur bloqué sur 0 ou sur 1
* Transition inattendues de signaux
L’ensemble de ces comportements est modélisé par un ensemble d’états discret :
* Un état de fonctionnement normal
* Plusieurs états de fonctionnement dégradés correspondant à des défauts spécifiques

# Conclusion 
Il était question pour nous dans ce travail de faire une preuve de concept de l'article scientifique portant sur une approche de diagnostic en ligne des défaillances des systèmes à évènement discret basée sur l'apprentissage automatique. Au bout de nous simulations, nous avons obtenu une accuracy de 99.13%, supérieure aux 70% de l'article. Cette différence s'explique par la simplicité relative de notre système simulée (3 composants contre 33 dans l’article). Notre implémentation valide l’approche méthodologique de Saddem et al (2022) et démontre l’efficacité des réseaux LSTM pour le diagnostic des défauts. Pour un système industriel réel plus complexe, nous nous attendrions a des performances similaires à celle de l’article.Nous notons cependant que bien que l'article parle d'un diagnostic en ligne, aucun développement n'est fait pour montrer comment le déploiement se fera, pour le traitement en temps réel des défaillances.
